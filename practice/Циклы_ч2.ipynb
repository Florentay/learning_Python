**Задача 1**

Переменная **car_speed** хранит скорость движения автомобиля. 

Переменная **speed_limit** — разрешённую скорость на текущем участке дороги.


Напишите программу, которая снизит скорость автомобиля до разрешённой в случае превышения.

```python
from random import randint

car_speed = 80
speed_limit = 60

print('Скорость автомобиля:', car_speed)
print('Разрешённая скорость:', speed_limit)

while car_speed > speed_limit:
    car_speed -= randint(1, 60)

print()
print('Скорость автомобиля:', car_speed)
print('Разрешённая скорость:', speed_limit)
```
