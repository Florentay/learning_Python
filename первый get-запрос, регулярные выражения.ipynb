```python
#регулярные выражения
#метод search(pattern, string) ищет шаблон pattern в строке srting
#хотя search () ищет шаблон во всей строке, возвращает он только первую найденную подстроку
import re
string = "Основы аналитики данных"
print(re.search('\w+', string))
```

```python
#метод search возвращает обьект типа match (англ. "соответствовать"), параметр span (англ диапазон) указывает диапазон индексов, подходящих под шаблон. В параметре match указано само значение подстроки 
#шаблону '\w+' соотвествует любая подстрока, содержащая одну и более букв, цифр или символ нижнего подчеркивания _. Метод search нашел, что этому шаблону соотвествует первое слово в строке. Так как под правило '\w+' не подходит пробел, метод вернул всё, что идет до первого пробела
```

```python
#если не нужны доп сведения о диапазоне
import re
string = "Основы аналитики данных"
print(re.search('\w+', string).group())
```

```python
import re
string = "Основы «аналитики данных»"
print(re.search('«[А-я ]+»', string).group())
```

```python
#метод split(pattern, string) разделяет строку string по границе шаблона pattern
import re
string = "Основы «аналитики данных»"
print(re.split('\d+', string))
```

```python
#метод findall() удобен тем что можно сразу посчитать количество повторяющихся подстрок в строке функцией len()
import re
string = "Основы «аналитики данных»"
print(len(re.findall('\w+', string)))
```
