7 видов питания 

```python
nomeal = 0 #без питания
breakfast = 0 #только завтрак 
full_board = 0  #FB (Full Board) обозначается трёхразовое питание — завтрак, обед и ужин (полный пансион)
dinner = 0 #только ужин 
lunch = 0 #только обед
half_board_dinner = 0 #завтрак и ужин (полупансион)
half_board_lunch = 0 #завтрак и обед (полупансион)


for i in range(len(data['data']['hotels'])): 
    for r in range(len(data['data']['hotels'][i]['rates'])):  
        if data['data']['hotels'][i]['rates'][r]['meal'] == 'nomeal':
            nomeal += 1
        elif data['data']['hotels'][i]['rates'][r]['meal'] == 'breakfast':
            breakfast += 1
        elif data['data']['hotels'][i]['rates'][r]['meal'] == 'full-board':
            full_board += 1
        elif data['data']['hotels'][i]['rates'][r]['meal'] == 'dinner':
            dinner += 1
        elif data['data']['hotels'][i]['rates'][r]['meal'] == 'lunch':
            lunch += 1
        elif data['data']['hotels'][i]['rates'][r]['meal'] == 'half-board-dinner':
            half_board_dinner += 1
        elif data['data']['hotels'][i]['rates'][r]['meal'] == 'half-board-lunch':
            half_board_lunch += 1

print(nomeal)
print(breakfast)
print(full_board)
print(dinner)
print(lunch)
print(half_board_dinner)
print(half_board_lunch)
```

```python
none = 0
time1 = 0 
time2 = 0
time3 = 0
time4 = 0


for i in range(len(data['data']['hotels'])): 
    for r in range(len(data['data']['hotels'][i]['rates'])):  
        for y in range(len(data['data']['hotels'][i]['rates'][r]['payment_options']['payment_types'])):
            if data['data']['hotels'][i]['rates'][r]['payment_options']['payment_types'][y]['cancellation_penalties']['free_cancellation_before'] == '2023-11-21T11:00:00':
                time1 += 1
            elif data['data']['hotels'][i]['rates'][r]['payment_options']['payment_types'][y]['cancellation_penalties']['free_cancellation_before'] == '2023-11-23T10:30:00':
                time2 += 1
            elif data['data']['hotels'][i]['rates'][r]['payment_options']['payment_types'][y]['cancellation_penalties']['free_cancellation_before'] == '2023-11-23T11:00:00':
                time3 += 1
            elif data['data']['hotels'][i]['rates'][r]['payment_options']['payment_types'][y]['cancellation_penalties']['free_cancellation_before'] == '2023-11-15T18:00:00':
                time4 += 1
            elif data['data']['hotels'][i]['rates'][r]['payment_options']['payment_types'][y]['cancellation_penalties']['free_cancellation_before'] == None:
                none += 1

print(f'Нет бесплатной отмены в {none} номерах')
print(f'Есть возможность бесплатной отмены до 11:00 часов 21 ноября 2023 в {time1} номерах')
print(f'Есть возможность бесплатной отмены до 10:30 часов 23 ноября 2023 в {time2} номерах')
print(f'Есть возможность бесплатной отмены до 11:00 часов 23 ноября 2023 в {time3} номерах')
print(f'Есть возможность бесплатной отмены до 18:00 часов 15 ноября 2023 в {time4} номерах')
```
